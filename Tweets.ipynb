```python
import re

def clean(tweet, location, time, geo):
    tweet = tweet.lower()
    tweet = re.sub("https?:[\d\w@_./]+", ' ', tweet)
    tweet = re.sub("[\n\t]", ' ', tweet)
    tweet = re.sub("#[\d\w:]+\s", ' ', tweet)
    tweet = re.sub("@[\d\w:]+\s", ' ', tweet)
    tweet = re.sub(r"what's", "what is ", tweet)
    tweet = re.sub(r"\'s", " is ", tweet)
    tweet = re.sub(r"\'ve", " have ", tweet)
    tweet = re.sub(r"can't", "cannot ", tweet)
    tweet = re.sub(r"n't", " not ", tweet)
    tweet = re.sub(r"i'm", "i am ", tweet)
    tweet = re.sub(r"\'re", " are ", tweet)
    tweet = re.sub(r"\'d", " would ", tweet)
    tweet = re.sub(r"\'ll", " will ", tweet)
    tweet = re.sub("[!.://\\@,'']", ' ', tweet)
    tweet = re.sub("[^a-zA-Z]", ' ', tweet)
    tweet = re.sub("\s+", ' ', tweet)
    return tweet, location, time, geo
```

```python
txt = [clean(x._json['text'], x._json['user']['location'], x._json['created_at'], x._json['geo']) for x in fetched_tweets]
```
