```python
def extract_articles():
    all_articles = newsapi.get_everything(q='cricket',
                                      sources='bbc-news,the-verge',
                                      domains='bbc.co.uk,techcrunch.com',
                                      from_param='2018-12-29',
                                      to='2018-04-29',
                                      language='en',
                                      sort_by='relevancy',
                                      page_size=100,
                                      page=1)
    headlines = [x['title'] for x in all_articles['articles']]
    contents = [x['content'] for x in all_articles['articles']]
    NEWS = [a+' '+b if b is not None else a for a,b in zip(headlines, contents)]
    total_articles =  all_articles['totalResults']
    last_ith = total_articles % 100 != 0
    for i in range(2, (total_articles // 100) + last_ith + 2):
        all_articles = newsapi.get_everything(q='cricket',
                                      sources='bbc-news,the-verge',
                                      domains='bbc.co.uk,techcrunch.com',
                                      from_param='2018-12-29',
                                      to='2018-04-29',
                                      language='en',
                                      sort_by='popularity',
                                      page_size=100,
                                      page=i)
        headlines = [x['title'] for x in all_articles['articles']]
        contents = [x['content'] for x in all_articles['articles']]
        NEWS = NEWS + [a+' '+b if b is not None else a for a,b in zip(headlines, contents)] 
    return NEWS
        
```
